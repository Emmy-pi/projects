### Predictive Analysis


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,classification_report,roc_curve,accuracy_score,auc


```

```python
df.head()
```

```python
df['Churn']=df.Churn.map({'No':0,'Yes':1})
```

```python
df.head()
```

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
df.sample(frac=.10)
```

```python
x_hot_encoder=pd.get_dummies(df)
```

```python
x_hot_encoder.head().T
```

```python
X=x_hot_encoder.drop('Churn', axis=1)
y=x_hot_encoder['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42, stratify = y)
```

```python
# Scaling all the variables to a range of 0 to 1
from sklearn.preprocessing import MinMaxScaler
features = X.columns.values
scaler = MinMaxScaler(feature_range = (0,1))
scaler.fit(X)
X = pd.DataFrame(scaler.transform(X))
X.columns = features
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
classifier=LogisticRegression(random_state=42)
classifier.fit(X_train,y_train)
Log_Score=classifier.score(X_test,y_test)*100
Log_Score
```

```python
classifier = RandomForestClassifier(n_estimators=250,random_state=15)
classifier.fit(X_train, y_train)
y_pred=classifier.predict(X_test)
RF_score=classifier.score(X_test,y_test)*100
RF_score
```

```python
gb = GradientBoostingClassifier()
gb.fit(X_train, y_train)
y_pred = gb.predict(X_test)
GB_score= (gb.score(X_test, y_test)*100)
GB_score
```

```python
clf = KNeighborsClassifier(n_neighbors=3) 
clf.fit(X_train,y_train)  
KNN_score=clf.score(X_test,y_test)*100
KNN_score
```

```python
model=SVC(kernel='rbf')                                 
model.fit(X_train,y_train)
svc_score=model.score(X_test,y_test)*100
svc_score
```

```python
models = ['SVC', 'RandomForestClassifier' , 'KNeighborsClassifier', 
          'GradientBoostingClassifier', 'LogisticRegression']
scores = [svc_score, RF_score ,KNN_score, GB_score, Log_Score]
score_table = pd.DataFrame({'Model':models, 'Score':scores})
score_table.sort_values(by='Score', axis=0, ascending=False)
print(score_table.sort_values(by='Score', ascending=False))
sns.barplot(x = score_table['Score'], y = score_table['Model'], palette='viridis');
```

## Observation
* Logistic Regression and Gradient Boosting Classifier provides better prediction accuracy.
